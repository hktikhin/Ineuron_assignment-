#### In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,'lxml')
```

```python
text = soup.get_text()
```

```python
text
```

```python
## tokenize text into sentence
sentences = nltk.sent_tokenize(text)
```

```python
## removing new line chracter 
clean_sentences = []
for sentence in sentences:
    for cleanSent in sentence.split('\n'):
        cleanSent = cleanSent.strip()
        if cleanSent:
            clean_sentences.append(cleanSent)
```

```python
clean_sentences
```

```python
## word tokenization
## removing stop word and punctuation
from string import punctuation
from nltk.corpus import stopwords
stopWords = stopwords.words('english')
words = []
for sent in clean_sentences:
    for word in nltk.word_tokenize(sent):
        word = word.lower()
        if word not in punctuation and word not in stopWords:
            words.append(word)
```

```python
## try to do some stemming --snowball stemming 
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer('english')
words = [snowball.stem(word) for word in words]
```

```python
from collections import Counter
Counter(words)
```

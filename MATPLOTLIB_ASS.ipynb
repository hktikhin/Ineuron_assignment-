## Promble 1
### We have the min and max temperatures in a city In India for each months of the year.We would like to find a function to describe this and show it graphically, the dataset given below.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
month = np.arange(1,13)
```

```python
# why show both in some axis?
plt.scatter(month,Min, c='b')
plt.scatter(month,Max, c='r')
```

```python
month
```

```python
# we want to define a function for line fitting 
from scipy.optimize import curve_fit
# fitting with sin function 
def func(x, a, b, c, d):
    return a * np.sin(b*x + c) + d
```

```python
popt1, pcov = curve_fit(func, month, Min)
```

```python
popt2, pcov = curve_fit(func, month, Max)
```

```python
popt1
```

```python
plt.scatter(month,Min, c='b')
plt.scatter(month,Max, c='r')
plt.plot(month, func(month, *popt1), 'b-')
plt.plot(month, func(month, *popt2), 'r-')
plt.xlabel('Month')
plt.ylabel('temperature (Â°c)')
```

## Problem 2
### 1. Create a pie chart presenting the male/female proportionã
### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
```

```python
ax = titanic["sex"].value_counts().plot.pie(figsize = (5,5),fontsize = 15, autopct = '%1.1f%%',shadow=True)
ax.set_ylabel('Sex',fontsize = 20)
```

```python
# what is cmap?
# How to plot in same graph? 
color = ['orange','blue']
fig, ax = plt.subplots(figsize=(10, 5))

for gender,c in zip(titanic["sex"].unique(),color):
    groupdata = titanic[titanic["sex"] == gender]
    groupdata.plot.scatter(y = 'fare', x = 'age',c = c,ax=ax,alpha =0.5,label=gender)
    
ax.legend()
```
